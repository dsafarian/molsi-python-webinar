## Tabular Data

```python
import numpy
import os
```

```python
#used to read in a file
help(numpy.genfromtxt)
```

```python
#delimiter: what we break up the file based on
#skip_headers: skip certain number of lines at the beginning
#dtype = unicode ; everything is words or strings
distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname = distance_file, delimiter = ',', dtype = 'unicode')
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
data = distances[1:]
print(data)
```

```python
#change data from string to int
data = data.astype(numpy.float)
print(data)
```

```python
#to slice 2D
#array_name[row_index, column_index]
print(data[0,1])
```

```python
small_data = data[0:10, 0:3]
print(small_data)
```

```python
print(small_data[:,1:])
```

```python
#data_average = numpy.mean(data_set)
print(headers)
```

```python
for col in range(1,len(headers)):
    col_data = data[:, col]
    avg = numpy.mean(col_data)
    print(F'{headers[col]}: {avg:.3f}')
```

## Plotting

```python
import matplotlib.pyplot as plt
#if you dont see plot in notebook
%matplotlib inline
```

```python
#to make new plot
plt.figure()
#can specify x & y, if don't input x, just adds values numerically (a count)
plt.plot(data[:,1])
```

```python
plt.figure()
sample_name = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:,1], label=sample_name)
#for label to show up
plt.legend()
#to save, specified resolution (dots per inch)
plt.savefig(F'{sample_name}.png', dpi=300)
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:,1], label = headers[1], color = 'red')
#to add another plot
plt.plot(data[:,2], label = headers[2])
#for label to show up
plt.legend()
#to save, specified resolution (dots per inch)
plt.savefig('twosamples.png', dpi=300)
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')

for col in range(1,len(headers)):
    plt.plot(data[:,col], label=headers[col])
    
plt.legend()
plt.savefig('all_samples.png', dpi=300)
```

```python
#to figure out min and max
data_min = numpy.min(data[:,1:])
data_max = numpy.max(data[:,1:])


for col in range(1, len(headers)):
    plt.figure()
    sample_name = headers[col]
    plt.ylim(data_min, data_max)
       
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.plot(data[:,col], label=sample_name)
    plt.legend()
    plt.savefig(F'{sample_name}.png', dpi=300)
```

```python
#can increment rows
print(data[:10:2, :])
```

```python
plt.figure()
#plt.plot(x,y)
plt.plot(data[:10:2, 0], data[:10:2, 1])
```

```python
#doesn't specify x value, only counts 100 data points
plt.figure()
plt.plot(data[::100,1])
```

```python
#line style before label
# '--', 'o', '-o', etc. 
plt.figure()
plt.plot(data[::100,0], data[::100,1], '-o', label = headers[1], color = 'orange')
plt.legend()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
```
